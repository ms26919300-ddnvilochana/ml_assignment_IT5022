```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set style for visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

# Define dataset paths
dataset_path = Path(r'd:\MSc in AI\Fundamentals of ML\Project01\ml_assignment_IT5022\dataset')
insurance_file = dataset_path / 'Motor vehicle insurance data.csv'
claim_file = dataset_path / 'sample type claim.csv'

# Load datasets
insurance_df = pd.read_csv(insurance_file, sep=';')
claim_df = pd.read_csv(claim_file, sep=';')

print(f"Insurance dataset: {insurance_df.shape}")
print(f"Claims dataset: {claim_df.shape}")
```

```python
# Insurance Dataset Overview

print("MOTOR VEHICLE INSURANCE DATA - SUMMARY")
print("-"*90)
print(f"\nColumn Names and Data Types:")
print(insurance_df.dtypes)

print(f"\n\nFirst Few Records:")
print(insurance_df.head(3))

print(f"\n\nMissing Values:")
missing_values = insurance_df.isnull().sum()
if missing_values.sum() > 0:
    print(missing_values[missing_values > 0])
else:
    print("No missing values")

print(f"\n\nBasic Statistics:")
print(insurance_df.describe())
```

```python
# Claims Dataset Overview

print("SAMPLE TYPE CLAIM DATA - SUMMARY")
print("-"*90)
print(f"\nColumn Names and Data Types:")
print(claim_df.dtypes)

print(f"\n\nFirst Few Records:")
print(claim_df.head(10))

print(f"\n\nMissing Values:")
missing_values = claim_df.isnull().sum()
if missing_values.sum() > 0:
    print(missing_values[missing_values > 0])
else:
    print("No missing values")

print(f"\n\nBasic Statistics:")
print(claim_df.describe())
```

```python
# Premium Analysis

print("PREMIUM ANALYSIS")
print("-"*90)
print(f"  Average Premium:        {insurance_df['Premium'].mean():.2f}")
print(f"  Median Premium:         {insurance_df['Premium'].median():.2f}")
print(f"  Min Premium:            {insurance_df['Premium'].min():.2f}")
print(f"  Max Premium:            {insurance_df['Premium'].max():.2f}")
print(f"  Std Dev:                {insurance_df['Premium'].std():.2f}")
print(f"  25th Percentile:        {insurance_df['Premium'].quantile(0.25):.2f}")
print(f"  75th Percentile:        {insurance_df['Premium'].quantile(0.75):.2f}")
print(f"  Interquartile Range:    {insurance_df['Premium'].quantile(0.75) - insurance_df['Premium'].quantile(0.25):.2f}")
```

```python
# Claims Analysis

print("2.2 CLAIMS ANALYSIS")
print("-"*90)
print(f"  Avg Claims Cost/Year:             {insurance_df['Cost_claims_year'].mean():.2f}")
print(f"  Avg Number of Claims/Year:        {insurance_df['N_claims_year'].mean():.2f}")
print(f"  Avg Number of Claims in History:  {insurance_df['N_claims_history'].mean():.2f}")

# Claim ratio
claims_ratio = (insurance_df['N_claims_year'] > 0).sum()
print(f"  Policyholders with Claims:        {claims_ratio:,} ({claims_ratio/len(insurance_df)*100:.1f}%)")

# Distribution of claims this year
print(f"\n  Distribution of Claims This Year:")
for n in range(0, 6):
    count = (insurance_df['N_claims_year'] == n).sum()
    pct = count / len(insurance_df) * 100
    print(f"    {n} claims: {count:,} ({pct:.2f}%)")
```

```python
# Policy Lapse (Cancellation) Analysis

print("POLICY LAPSE & RETENTION ANALYSIS")
print("-"*90)
total_policies = len(insurance_df)
lapsed = insurance_df['Lapse'].sum()
active = (insurance_df['Lapse'] == 0).sum()

print(f"  Total Policies:          {total_policies:,}")
print(f"  Lapsed Policies:         {lapsed:,} ({lapsed/total_policies*100:.2f}%)")
print(f"  Active Policies:         {active:,} ({active/total_policies*100:.2f}%)")
print(f"  Retention Rate:          {active/total_policies*100:.2f}%")
print(f"  Churn Rate:              {lapsed/total_policies*100:.2f}%")
```

```python
# Vehicle Characteristics

print("2.4 VEHICLE CHARACTERISTICS")
print("-"*90)
print(f"  Average Power:                {insurance_df['Power'].mean():.1f} units")
print(f"  Average Cylinder Capacity:    {insurance_df['Cylinder_capacity'].mean():.1f}")
print(f"  Average Vehicle Value:        {insurance_df['Value_vehicle'].mean():.2f}")
print(f"  Average Vehicle Age (2025):   {2025 - insurance_df['Year_matriculation'].mean():.1f} years")
print(f"  Average Weight:               {insurance_df['Weight'].mean():.1f} kg")
print(f"  Average Length:               {insurance_df['Length'].mean():.3f} meters")
print(f"\n  Fuel Type Distribution:")

fuel_counts = insurance_df['Type_fuel'].value_counts()
for fuel, count in fuel_counts.items():
    pct = count / len(insurance_df) * 100
    print(f"    {fuel}: {count:,} ({pct:.2f}%)")

print(f"\n  Number of Doors Distribution:")
doors_counts = insurance_df['N_doors'].value_counts().sort_index()
for doors, count in doors_counts.items():
    pct = count / len(insurance_df) * 100
    print(f"    {doors} doors: {count:,} ({pct:.2f}%)")
```

```python
# Customer & Distribution Analysis

print("2.5 CUSTOMER & DISTRIBUTION ANALYSIS")
print("-"*90)

# Seniority
print(f"  Customer Seniority (Years with Company):")
print(f"    Average:              {insurance_df['Seniority'].mean():.1f} years")
print(f"    Median:               {insurance_df['Seniority'].median():.1f} years")
print(f"    Min:                  {insurance_df['Seniority'].min()} years")
print(f"    Max:                  {insurance_df['Seniority'].max()} years")

# Distribution Channel
print(f"\n  Distribution Channel:")
dist_counts = insurance_df['Distribution_channel'].value_counts()
for channel, count in dist_counts.items():
    pct = count / len(insurance_df) * 100
    print(f"    Channel {channel}: {count:,} ({pct:.2f}%)")

# Payment Method
print(f"\n  Payment Method Distribution:")
pay_counts = insurance_df['Payment'].value_counts()
for method, count in pay_counts.items():
    pct = count / len(insurance_df) * 100
    print(f"    Method {method}: {count:,} ({pct:.2f}%)")

# Second Driver
second_driver_yes = (insurance_df['Second_driver'] == 1).sum()
second_driver_no = (insurance_df['Second_driver'] == 0).sum()
print(f"\n  Second Driver:")
print(f"    Yes: {second_driver_yes:,} ({second_driver_yes/len(insurance_df)*100:.2f}%)")
print(f"    No:  {second_driver_no:,} ({second_driver_no/len(insurance_df)*100:.2f}%)")
```

```python
# Claims Dataset Analysis

print("2.6 CLAIMS DATASET ANALYSIS")
print("-"*90)

print(f"  Total Claims Records:            {len(claim_df):,}")
print(f"  Unique Customers with Claims:    {claim_df['ID'].nunique():,}")
print(f"  Coverage Rate:                   {claim_df['ID'].nunique()/len(insurance_df)*100:.2f}%")
print(f"  Customers with Multiple Claims:  {(claim_df['ID'].value_counts() > 1).sum():,}")
print(f"  Avg Claims per Customer:         {len(claim_df)/claim_df['ID'].nunique():.2f}")

print(f"\n  Claims Type Distribution:")
claim_type_dist = claim_df['Claims_type'].value_counts()
for claim_type, count in claim_type_dist.items():
    pct = count / len(claim_df) * 100
    print(f"    {claim_type:20s}: {count:5,} ({pct:5.2f}%)")

print(f"\n  Cost Analysis by Claim Type:")
for claim_type in claim_df['Claims_type'].unique():
    subset = claim_df[claim_df['Claims_type'] == claim_type]
    avg_total = subset['Cost_claims_year'].mean()
    avg_type = subset['Cost_claims_by_type'].mean()
    total_cost = subset['Cost_claims_by_type'].sum()
    print(f"    {claim_type:20s}: Avg Type Cost {avg_type:8.2f}, Total {total_cost:12,.2f}")
```

```python
# Summary Statistics

summary_data = {
    'Metric': [
        'Total Policies',
        'Active Policies',
        'Lapsed Policies',
        'Unique Claimants',
        'Average Premium',
        'Claims Incidence Rate',
        'Average Claims Cost',
        'Average Policy Tenure',
        'Retention Rate'
    ],
    'Value': [
        f'{len(insurance_df):,}',
        f'{(insurance_df["Lapse"] == 0).sum():,}',
        f'{insurance_df["Lapse"].sum():,}',
        f'{claim_df["ID"].nunique():,}',
        f'{insurance_df["Premium"].mean():.2f}',
        f'{(insurance_df["N_claims_year"] > 0).sum()/len(insurance_df)*100:.1f}%',
        f'{insurance_df["Cost_claims_year"].mean():.2f}',
        f'{insurance_df["Seniority"].mean():.1f} years',
        f'{(insurance_df["Lapse"] == 0).sum()/len(insurance_df)*100:.2f}%'
    ]
}

summary_df = pd.DataFrame(summary_data)
print("SUMMARY STATISTICS")
print("-"*90)
print(summary_df.to_string(index=False))
```
